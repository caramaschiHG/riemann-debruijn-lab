# Robin and Lagarias finite scans

This notebook runs a small finite scan using the package API. It does not prove RH; it only demonstrates the finite-computation workflow.

```python
from pathlib import Path
import sys

root = Path.cwd()
if not (root / "src" / "riemann_lab").exists():
    root = root.parent
sys.path.insert(0, str(root / "src"))

from riemann_lab.robin.divisor_sums import sigma_sieve
from riemann_lab.robin.lagarias_scan import scan_lagarias
from riemann_lab.robin.robin_scan import scan_robin

limit = 10_000
sigma = sigma_sieve(limit)
robin_rows = scan_robin(limit, top=5, sigma=sigma)
lagarias_rows = scan_lagarias(limit, top=5, sigma=sigma)

{
    "robin_top_n": [row.n for row in robin_rows],
    "lagarias_top_n": [row.n for row in lagarias_rows],
}
```
