# Candidate comparison demo

This notebook inspects the committed candidate-comparison CSV and plots two finite diagnostics. It does not add an experiment, prove RH, or establish `Lambda <= 0`.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

root = Path.cwd()
if not (root / "outputs" / "candidate_comparison").exists():
    root = root.parent

csv_path = root / "outputs" / "candidate_comparison" / "candidate_comparison.csv"
frame = pd.read_csv(csv_path)

ranking_columns = [
    "candidate_id",
    "normalized_gap_metric",
    "explicit_gbar_upper",
    "collision_time_abs",
    "rank_normalized_gap",
    "rank_gbar_upper",
    "rank_collision_time",
]

frame.sort_values("rank_normalized_gap")[ranking_columns].head(8)
```

```python
plot_frame = frame.dropna(subset=["explicit_gbar_upper"])

fig, ax = plt.subplots(figsize=(7, 4))
ax.scatter(plot_frame["normalized_gap_metric"], plot_frame["explicit_gbar_upper"])
ax.set_xlabel("normalized gap")
ax.set_ylabel("explicit-style gbar upper")
ax.set_title("Finite candidate comparison")

for _, row in plot_frame.iterrows():
    if bool(row.get("candidate_a", False)):
        ax.annotate("Candidate A", (row["normalized_gap_metric"], row["explicit_gbar_upper"]))

fig.tight_layout()
plt.show()
```
